```python
# Initialize empty dictionary
rename_dict = {}
# List of wanted names
wanted_names = ['1.2: Capital accumulation',
                '1.3: Capital expenses',
                '1.4: Current and capital expenditure (1+3)',
                '1.1: Current expenditure',
                '2.1: Current revenue',
                '2.2: Capital revenue',
                '2.3: Current plus capital revenue (1+2)',
                '2.4: Currents surplus=Gross saving (2.1-1.1)',
                '2.5: Overall surplus=Net lending/borrowing (2.3-1.4)']
# Create rename dictionary for variable names
for name, wantedname in zip(names,wanted_names):
    rename_dict[name] = wantedname
# Rename the variables
for key,value in rename_dict.items():
    OFF3.variable.replace(key, value, inplace=True)
```

```python
names = ['capital_accumulation', 'capital_expenses', 'current_and_capital_expenditure',
                 'current_expenditure', 'current_revenue', 'capital_revenue', 'current_plus_capital_revenue',
                 'current_surplus', 'overall_surplus']
for i, name in enumerate(names):
    namespace = locals()
    namespace[name] = (OFF3[Ilist[i]])
```

```python
dataframenames = [capital_accumulation, capital_expenses, current_and_capital_expenditure,
                 current_expenditure, current_revenue, capital_revenue, current_plus_capital_revenue,
                 current_surplus, overall_surplus]
#for i in range(0,9):
 #   dataframenames[i] = dataframenames[i].reset_index(drop=True, inplace=True)
for i in dataframenames:
    i.reset_index(drop=True, inplace=True)
```

```python
capital_expenses['Diff'] = capital_expenses['mio. DKK'].diff()
```

```python
capital_expenses[:3]
```

```python
capital_expenses['Avg. abs. change'] = capital_expenses['Diff'].mean()
```

```python
capital_expenses[:3]
```

```python
capital_expenses['Pct. change'] = capital_expenses['mio. DKK'].pct_change()*100
```

```python
capital_expenses[:3]
```

```python
capital_expenses['Avg. pct. change'] = ((capital_expenses['mio. DKK'].iloc[-1]/capital_expenses['mio. DKK'][0])**(1/len(capital_expenses['mio. DKK']))-1)*100
```

```python
capital_expenses[:3]
```

```python
capital_expenses['Change in pct. points'] = capital_expenses['Pct. change'].diff()
```

```python
capital_expenses[:3]
```

```python
capital_expenses['Simple index'] = np.nan
for i in range(0,len(capital_expenses['mio. DKK'])):
    capital_expenses['Simple index'][i] = capital_expenses['mio. DKK'].iloc[i]/capital_expenses['mio. DKK'].iloc[0]*100
```

# Interactive plotting

```python
import ipywidgets as widgets
from ipywidgets import interact
 
#items = ['All']+sorted(df['Asset Type Description'].unique().tolist())

rows = ['mio. DKK', 'Diff', 'Avg. abs. change', 'Pct. change', 'Avg. pct. change', 'Change in pct. points', 'Simple index' ]
 
def view(x):
    plt.plot(capital_expenses['time'], capital_expenses[x])
    plt.ylabel(x)
    return plt.show()


def view2(x, df):
    plt.plot(dataframenames[df]['time'], dataframenames[df][x])
    plt.ylabel(x)
    return plt.show()
w = widgets.Select(options=rows)
q = widgets.Select(options=range(0,9),value=1)
interact(view2, x=w,df=q)
```
